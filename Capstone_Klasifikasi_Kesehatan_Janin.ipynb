# **Data Processing**

**Feature Selection**

Seleksi fitur harus dilakukan untuk menemukan rangkaian fitur terbaik yang memungkinkan terciptanya model optimal dari fenomena yang dipelajari.

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = data.drop(['fetal_health'], axis=1)
Y = data['fetal_health']

bestfeatures = SelectKBest(score_func=f_classif, k='all')

fit = bestfeatures.fit(X, Y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
# Concatenate two dataframes for better understanding
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Feature', 'Score']

# Sort features based on importance scores
featureScores = featureScores.sort_values(by='Score', ascending=False)
print(featureScores)
```

```python
# Visualize the importance scores
fig, ax = plt.subplots(figsize=(15, 10))
plot = sns.barplot(data=featureScores, x='Score', y='Feature', palette='coolwarm', linewidth=0.5, saturation=2,
                   orient='h')

plt.title('Feature Importance',
          fontsize=24,
          fontweight='bold')
```

```python
# Finding the least correlated features with fetal health
low_score_features = featureScores[featureScores['Score'] < 15]
print(low_score_features)
```

```python
# Dropping the features that are below the threshold
features_to_drop = ['histogram_number_of_peaks', 'fetal_movement', 'histogram_max', 'histogram_number_of_zeroes']
data = data.drop(features_to_drop, axis=1)
```

**Feature Scaling**

Penskalaan Fitur adalah teknik untuk menstandardisasi fitur-fitur individual dalam data ke dalam rentang yang tetap. Teknik ini dilakukan untuk menangani kuantitas, nilai, atau unit yang sangat bervariasi.

Jika penskalaan fitur tidak dilakukan, algoritma pembelajaran mesin akan cenderung memberi bobot pada nilai yang lebih besar sebagai nilai yang lebih tinggi dan memperlakukan nilai yang lebih kecil sebagai nilai yang lebih rendah, terlepas dari unit nilai tersebut.

```python
# Ranges of the features
plt.figure(figsize=(20, 10))
sns.boxenplot(data=data, palette='coolwarm')

plt.title('Ranges of the Features',
          fontsize=24,
          fontweight='bold')
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=12)
```

Value ranges of the features are very different and performing operations with these values will not lead to a healthy result. Therefore, features need to be scaled for a certain range.

```python
# Splitting data into features (X) and outcomes (y)
X = data.drop(["fetal_health"], axis=1)
y = data["fetal_health"]
```

**Standard Scaler**

```python
from sklearn import preprocessing

# Set up a standard scaler for the features
col_names = list(X.columns)
s_scaler = preprocessing.StandardScaler()

X_df = s_scaler.fit_transform(X)
X_df = pd.DataFrame(X_df, columns=col_names)
```

```python
# Description of the standardized features
X_df.describe().T
```

```python
# Plot the standardized features
plt.figure(figsize=(20, 10))
sns.boxenplot(data=X_df, palette='coolwarm')

plt.title('Ranges of the Standardized Features',
          fontsize=24,
          fontweight='bold')
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=12)
```

Numerical input values in the dataset were standardized in the pre-processing step with removing the mean and scaling each feature to unit variance to express correlations in training the models as can be seen in the figure above.

# **Model Evalution**

**Splitting the dataset into training and test test**

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test set
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
from tensorflow.keras.utils import to_categorical

y_train_encoded = to_categorical(y_train - 1, num_classes=3)


class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)


class_weight_dict = {i + 1: weight for i, weight in enumerate(class_weights)}
print("Class weights:", class_weight_dict)

y_test_categorical = to_categorical(y_test - 1, num_classes=3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(3, activation='softmax'))  # 3 kelas output

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(
    X_train, y_train_encoded,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    class_weight=class_weight_dict,
    verbose=1
)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test_categorical)
print(f'Test loss: {loss:.4f}')
print(f'Test accuracy: {accuracy:.4f}')

```

```python
# Plot training history
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
